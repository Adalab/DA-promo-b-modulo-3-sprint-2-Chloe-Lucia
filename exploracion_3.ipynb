```python
def tolist(x):
    if len(x.split(',')) == 2:
        lista = [x.split(',')[0], x.split(',')[1]]
        return lista
    else:
        lista2 = [x.split(',')[0], x.split(',')[1], x.split(',')[2]]
        return lista2
```

```python
state['State'] = state['State'].apply(tolist)
```
